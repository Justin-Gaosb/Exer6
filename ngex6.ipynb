Some typical NumPy 'random' functions:

```python
import numpy as np

# To generate a random float values between 0 and 1 (exclusive)
print(np.random.rand(3, 2)) # Creates a 3 x 2 matrix of random float values between 0 and 1 (exclusive)

# To generate random float values from standard normal distribution (mean = 0 and std = 1)
print(np.random.randn(5)) # Generates 5 random float values from standard normal distribution

# To generate random integer values within a range of values
print(np.random.randint(1, 100, 10)) # Generates 10 random integer values between 1 and 100 (exclusive)

# To randomly select an element from a given list of elements
print(np.random.choice([1, 2, 3, 4, 5, 6, 7, 8, 9], size=3)) # Three random numbers from the given array will be chosen

# To generate a random sample of values from a normally distributed with a specified mean and standard deviation
print(np.random.normal(loc=0, scale=1, size=10)) # 10 random values will be generated from a normally distributed data where 0 (loc) is the mean and 1 (scale) is the standard deviation

# To set a seed for a random number generation to ensure you get the same results later as well
print(np.random.seed(42)) # This ensures that any random numbers generated in the program are same every time you run the program. The seed value '42' can be any other value
```

AIM #1: Generate a very large dataset
1. Generate a dataset of 1 million random data items between 1 and 100 items using only pandas
2. Generate a dataset of 1 million random data items between 1 and 100 using only NumPy
3. Calculate the time it takes for both the above operations. 
    3.1. Import the 'time' module, and use the time() function to calculate current time
    3.2. Which one is faster and why?

```python
import pandas as pd
import numpy as np
import time

def measure_time_pandas():
    start_time = time.time()
    df_pandas = pd.DataFrame({'data': range(1, 1000001)})
    df_pandas['random_data'] = df_pandas['data'].apply(lambda x: np.random.randint(1, 101))
    end_time = time.time()
    return end_time - start_time
def measure_time_numpy():
    start_time = time.time()
    np_random_data = np.random.randint(1, 101, size=1000000)
    end_time = time.time()
    return end_time - start_time

pandas_time = measure_time_pandas()
print(f"Pandas time taken: {pandas_time} seconds")

numpy_time = measure_time_numpy()
print(f"NumPy time taken: {numpy_time} seconds")

if pandas_time < numpy_time:
    print("Pandas is faster.")
else:
    print("NumPy is faster.")
```

AIM #2: Basic statistics
For the given dataset on sleep health and lifestyle, do the following
1. Using only pandas, load the dataset, calculate mean 'Sleep Duration', 'Systolic Blood Pressure', 'Diastolic Blood Pressure', 'Heart Rate' and 'Daily Steps'.
2. Do the same as in Step 1 using only NumPy
3. Using only pandas, first calculate correlation (across only the numerical variables), and then separate correlation between...
    Sleep duration and Age
    Sleep duration and Heart rate
    Sleep duration and Daily steps
4. Using only NumPy, do the same as Step 3
5. Using pandas only, calculate standard deviation for 'Sleep Duration'. 
6. Usiong NumPy only, calculate standard deviation for 'Sleep Duration'. 
7. Calculate the time difference between using pandas and NumPy, right from the step of loading the dataset to the final standard deviation step. 
    5.1. Which one is faster and why?

```python
import pandas as pd
import numpy as np
import time

start_time_pandas = time.time()
df = pd.read_csv('sleep_health.csv')

mean_sleep_duration_pd = df['Sleep Duration'].mean()
mean_systolic_pd = df['Systolic blood pressure'].mean()
mean_diastolic_pd = df['Diastolic blood pressure'].mean()
mean_heart_rate_pd = df['Heart Rate'].mean()
mean_daily_steps_pd = df['Daily Steps'].mean()

numeric_cols = df.select_dtypes(include=[np.number]).columns
correlation_matrix_pd = df[numeric_cols].corr()
sleep_duration_age_corr_pd = correlation_matrix_pd['Sleep Duration']['Age']
sleep_duration_heart_rate_corr_pd = correlation_matrix_pd['Sleep Duration']['Heart Rate']
sleep_duration_daily_steps_corr_pd = correlation_matrix_pd['Sleep Duration']['Daily Steps']

std_sleep_duration_pd = df['Sleep Duration'].std()
time_pandas = time.time() - start_time_pandas

start_time_numpy = time.time()
data = np.genfromtxt('sleep_health.csv', delimiter=',', skip_header=1, dtype=None, encoding='utf-8', names=True)

print("Column names in CSV:", data.dtype.names)


sleep_duration_np = data['Sleep Duration'].astype(float)
systolic_blood_pressure_np = data['Systolic blood pressure'].astype(float)
diastolic_blood_pressure_np = data['Diastolic blood pressure'].astype(float)
heart_rate_np = data['Heart Rate'].astype(float)
daily_steps_np = data['Daily Steps'].astype(float)

mean_sleep_duration_np = np.mean(sleep_duration_np)
mean_systolic_np = np.mean(systolic_blood_pressure_np)
mean_diastolic_np = np.mean(diastolic_blood_pressure_np)
mean_heart_rate_np = np.mean(heart_rate_np)
mean_daily_steps_np = np.mean(daily_steps_np)
age_np = data['Age'].astype(float) 
sleep_duration_age_corr_np = np.corrcoef(sleep_duration_np, age_np)[0, 1]
sleep_duration_heart_rate_corr_np = np.corrcoef(sleep_duration_np, heart_rate_np)[0, 1]
sleep_duration_daily_steps_corr_np = np.corrcoef(sleep_duration_np, daily_steps_np)[0, 1]
std_sleep_duration_np = np.std(sleep_duration_np)
time_numpy = time.time() - start_time_numpy
print("Pandas Results:")
print(f"Mean Sleep Duration: {mean_sleep_duration_pd}")
print(f"Mean Systolic Blood Pressure: {mean_systolic_pd}")
print(f"Mean Diastolic Blood Pressure: {mean_diastolic_pd}")
print(f"Mean Heart Rate: {mean_heart_rate_pd}")

print(f"Mean Daily Steps: {mean_daily_steps_pd}")
print(f"Correlation (Sleep Duration vs Age): {sleep_duration_age_corr_pd}")
print(f"Correlation (Sleep Duration vs Heart Rate): {sleep_duration_heart_rate_corr_pd}")
print(f"Correlation (Sleep Duration vs Daily Steps): {sleep_duration_daily_steps_corr_pd}")
print(f"Standard Deviation of Sleep Duration: {std_sleep_duration_pd}")

print("\nNumPy Results:")
print(f"Mean Sleep Duration: {mean_sleep_duration_np}")
print(f"Mean Systolic Blood Pressure: {mean_systolic_np}")
print(f"Mean Diastolic Blood Pressure: {mean_diastolic_np}")
print(f"Mean Heart Rate: {mean_heart_rate_np}")
print(f"Mean Daily Steps: {mean_daily_steps_np}")
print(f"Correlation (Sleep Duration vs Age): {sleep_duration_age_corr_np}")
print(f"Correlation (Sleep Duration vs Heart Rate): {sleep_duration_heart_rate_corr_np}")
print(f"Correlation (Sleep Duration vs Daily Steps): {sleep_duration_daily_steps_corr_np}")
print(f"Standard Deviation of Sleep Duration: {std_sleep_duration_np}")

print(f"\nPandas Time: {time_pandas} seconds")
print(f"NumPy Time: {time_numpy} seconds")

if time_pandas < time_numpy:
    print("Pandas is faster.")
else:
    print("NumPy is faster.")
```

    5.4. Sleep Duration and Daily Steps

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time

df = pd.read_csv('sleep_health.csv')
start_time_pandas = time.time()
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
sns.histplot(df['Sleep Duration'], kde=True)
plt.title('Sleep Duration Distribution')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')
plt.show()

sns.histplot(df['Quality of Sleep'], kde=True)
plt.title('Quality of Sleep Distribution')
plt.xlabel('Quality of Sleep')
plt.ylabel('Frequency')
plt.show()

sns.histplot(df['Physical Activity Level'], kde=True)
plt.title('Physical Activity Level Distribution')
plt.xlabel('Physical Activity Level')
plt.ylabel('Frequency')
plt.show()

sns.histplot(df['Stress Level'], kde=True)
plt.title('Stress Level Distribution')
plt.xlabel('Stress Level')
plt.ylabel('Frequency')
plt.show()

sns.histplot(df['Heart Rate'], kde=True)
plt.title('Heart Rate Distribution')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Frequency')
plt.show()

time_pandas = time.time() - start_time_pandas

start_time_numpy = time.time()
age_unique, age_counts = np.unique(df['Age'].astype(int), return_counts=True)
plt.figure(figsize=(10, 6))
plt.hist(age_unique, bins=range(age_unique.min(), age_unique.max() + 2), weights=age_counts / age_counts.sum(), density=True, alpha=0.6, color='b')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

time_numpy = time.time() - start_time_numpy

categories = ['Quality of Sleep', 'Stress Level', 'Physical Activity Level', 'Occupation', 'BMI Category']
for category in categories:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=category, y='Sleep Duration', data=df)
    plt.title(f'Sleep Duration by {category}')
    plt.xlabel(category)
    plt.ylabel('Sleep Duration (hours)')
    plt.show()


for category in categories:
    cat_unique, cat_counts = np.unique(df[category], return_counts=True)
    sleep_duration_by_cat = [df[df[category] == i]['Sleep Duration'].values for i in cat_unique]
    plt.figure(figsize=(10, 6))
    plt.boxplot(sleep_duration_by_cat)
    plt.title(f'Sleep Duration by {category}')
    plt.xlabel(category)
    plt.ylabel('Sleep Duration (hours)')
    plt.xticks(ticks=range(1, len(cat_unique) + 1), labels=cat_unique)
    plt.show()


relations = [('Age', 'Sleep Duration'), ('Sleep Duration', 'Heart Rate'), ('Heart Rate', 'Daily Steps'), ('Sleep Duration', 'Daily Steps')]
for var1, var2 in relations:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=var1, y=var2, data=df)
    plt.title(f'Relation between {var1} and {var2}')
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.show()


for var1, var2 in relations:
    var1_unique, var1_counts = np.unique(df[var1].astype(float), return_counts=True)
    var2_unique, var2_counts = np.unique(df[var2].astype(float), return_counts=True)
    sleep_duration_by_var1 = [df[df[var1].astype(float) == i][var2].values for i in var1_unique]
    plt.figure(figsize=(10, 6))
    plt.scatter(var1_unique, var2_unique, alpha=0.6)
    plt.title(f'Relation between {var1} and {var2}')
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.show()
time_difference = time_pandas - time_numpy
print(f"Pandas Plotting Time: {time_pandas} seconds")
print(f"NumPy Plotting Time: {time_numpy} seconds")
print(f"Time Difference: {time_difference} seconds")

if time_pandas < time_numpy:
    print("Pandas is faster for plotting.")
else:
    print("NumPy is faster for plotting.")
```
